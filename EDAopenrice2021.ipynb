```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import re

plt.rcParams["figure.figsize"] = (20,15)
```

```python
import matplotlib as mpl
font_name = "STKaiti"
mpl.rcParams['font.family']=font_name
mpl.rcParams['axes.unicode_minus']=False # in case minus sign is shown as box
```

```python
df = pd.read_csv("./data/openrice2021 - Copy.csv",encoding='utf-8')
dfold = pd.read_csv("./data/open-rice.csv")

df = df.drop([7694])
```

```python
#lower case all names
df.Name = df.Name.apply(lambda x: x.lower() if isinstance(x, str) else x)
df.Name2 = df.Name2.apply(lambda x: x.lower() if isinstance(x, str) else x)
```

```python
df["Likes"] = df.Emoji.apply(lambda x: re.search("(\d+)",x.split(',')[0]).group(1))
df["Ok"] = df.Emoji.apply(lambda x: re.search("(\d+)",x.split(',')[1]).group(1))
df["Dislikes"] = df.Emoji.apply(lambda x: re.search("(\d+)",x.split(',')[2]).group(1))
```

```python
x = df.isnull()

sns.heatmap(data = x, cmap="viridis",cbar=False);
```

```python
df.District.value_counts()[:10].plot.bar()
plt.tick_params(axis='both', which='both', labelsize=24)
plt.xticks(rotation=0)
plt.ylabel("Count", fontdict={'fontsize': 32})
plt.title("Count of Restaurants per District (Top 10)", fontdict={'fontsize': 32})
plt.savefig("./data/District_Counts.png")
```
